```python
import altair as alt
from vega_datasets import data
```

```python
source = data.seattle_weather()

source
```

```python
chart1 = alt.Chart(source).mark_bar(
    cornerRadiusTopLeft=5,
    cornerRadiusTopRight=5
).encode(
    x=alt.X('weather:N', sort='-y'),
    y='count():Q',
    color='weather:N'
).properties(
    title="Кількість днів кожного типу погоди"
)

chart1
```

```python
source = data.barley()

source
```

```python
chart2 = alt.Chart(source).transform_filter(
    alt.datum.year == 1931
).mark_bar().encode(
    x='yield:Q',
    y=alt.Y('variety:N', sort='-x'),
    color='site:N'
).properties(
    title="Врожайність ячменю у 1931 році по сортах"
)

chart2
```

```python
source = data.seattle_weather()

chart3 = alt.Chart(source).mark_bar().encode(
    x=alt.X('month(date):O', title='Місяць'),
    y='count():Q',
    color='weather:N',
    column='weather:N'
).properties(
    title="Кількість днів кожного типу погоди по місяцях"
)

chart3
```

```python
source = data.stocks()

filtered = source[source['symbol'].isin(['GOOG', 'AAPL'])]

chart4 = alt.Chart(filtered).mark_area(
    line=True,
    color=alt.Gradient(
        gradient='linear',
        stops=[
            alt.GradientStop(color='white', offset=0),
            alt.GradientStop(color='steelblue', offset=1)
        ],
        x1=1, x2=1, y1=1, y2=0
    )
).encode(
    x='date:T',
    y='price:Q',
    color='symbol:N'
).properties(
    title="Ціни акцій GOOG і AAPL з градієнтом"
)

chart4
```

```python
source = data.unemployment_across_industries.url

chart5 = alt.Chart(source).mark_area().encode(
    x=alt.X('yearmonth(date):T', axis=alt.Axis(format='%Y')),
    y=alt.Y('sum(count):Q', stack='center', axis=None),
    color=alt.Color('series:N', scale=alt.Scale(scheme='category20b'))
).interactive().properties(
    width=700,
    height=400,
    title="Рівень безробіття в США по галузях"
)

chart5
```

```python
source = data.seattle_weather()

weather_counts = source.groupby('weather').size().reset_index(name='count')

chart6 = alt.Chart(weather_counts).mark_arc(innerRadius=50).encode(
    theta='count:Q',
    color='weather:N'
).properties(
    title="Типи погоди у Сіетлі — кільцева діаграма"
)

chart6
```

Додайте графік на вибір (інтерактивний)

```python
import altair as alt
from vega_datasets import data

source = data.cars()

origin_param = alt.param(
    name='origin_filter',
    value='USA',
    bind=alt.binding_select(
        options=['USA', 'Europe', 'Japan'],
        name='Походження авто: '
    )
)

chart7 = alt.Chart(source).mark_circle(size=90).encode(
    x=alt.X('Horsepower:Q', title='Потужність (Horsepower)'),
    y=alt.Y('Miles_per_Gallon:Q', title='Економність (Miles per Gallon)'),
    color=alt.Color('Origin:N', title='Регіон'),
    tooltip=[
        alt.Tooltip('Name:N', title='Модель'),
        alt.Tooltip('Origin:N', title='Походження'),
        alt.Tooltip('Horsepower:Q', title='Потужність'),
        alt.Tooltip('Miles_per_Gallon:Q', title='MPG'),
        alt.Tooltip('Year:T', title='Рік')
    ]
).add_params(
    origin_param
).transform_filter(
    alt.datum.Origin == origin_param
).properties(
    width=700,
    height=400,
    title='Інтерактивний графік: залежність потужності та економності авто'
).interactive()

chart7
```
