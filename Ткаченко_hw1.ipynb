```python
import numpy as np

# Для відтворюваності результатів
np.random.seed(42)
```

```python
# 1. Створіть одновимірний масив (вектор) з першими 20-ма натуральними числами та виведіть його значення.
vector_20 = np.arange(1, 21)
print('Вектор з перших 20 натуральних чисел:')
print(vector_20)
```

```python
# 2. Створіть двовимірний масив (матрицю) розміром 4x4, заповніть його одиницями та виведіть його значення.
matrix_ones = np.ones((4, 4), dtype=int)
print('Матриця 4x4, заповнена одиницями:')
print(matrix_ones)
```

```python
# 3. Створіть масив розміром 5x5, заповніть його випадковими дійсними числами в діапазоні від -1 до 1 та виведіть його значення.
matrix_rand_neg1_1 = np.random.uniform(-1, 1, (5, 5))
print('Масив 5x5 з випадковими дійсними числами від -1 до 1:')
print(matrix_rand_neg1_1)
```

```python
# 4. Створіть масив розміром 3x3, заповніть його випадковими дійсними числами в діапазоні від 0 до 1 та виведіть його значення.
matrix_rand_0_1 = np.random.rand(3, 3)
print('Масив 3x3 з випадковими дійсними числами від 0 до 1:')
print(matrix_rand_0_1)
```

```python
# 5. Створіть два одновимірних масиви розміром 6, заповніть їх випадковими цілими числами в діапазоні від 1 до 20 та виконайте на них поелементні операції додавання, віднімання та множення.
arr1 = np.random.randint(1, 21, 6)
arr2 = np.random.randint(1, 21, 6)
print('Перший масив:')
print(arr1)
print('Другий масив:')
print(arr2)
print('Поелементне додавання:')
print(arr1 + arr2)
print('Поелементне віднімання:')
print(arr1 - arr2)
print('Поелементне множення:')
print(arr1 * arr2)
```

```python
# 6. Створіть дві матриці розміром 3x2 та 2x4, заповніть їх випадковими цілими числами в діапазоні від 1 до 5 та перемножте їх між собою.
matrix_a = np.random.randint(1, 6, (3, 2))
matrix_b = np.random.randint(1, 6, (2, 4))
matrix_product_ab = np.dot(matrix_a, matrix_b)
print('Матриця 3x2:')
print(matrix_a)
print('Матриця 2x4:')
print(matrix_b)
print('Добуток матриць:')
print(matrix_product_ab)
```

```python
# 7. Створіть матрицю розміром 3x3, заповніть її випадковими дійсними числами в діапазоні від 0 до 1 та транспонуйте її.
matrix_transpose_source = np.random.rand(3, 3)
matrix_transposed = matrix_transpose_source.T
print('Початкова матриця 3x3:')
print(matrix_transpose_source)
print('Транспонована матриця:')
print(matrix_transposed)
```

```python
# 8. Створіть дві матриці розміром 3x3, заповніть їх випадковими цілими числами в діапазоні від 1 до 10 та знайдіть їхню різницю.
matrix_c = np.random.randint(1, 11, (3, 3))
matrix_d = np.random.randint(1, 11, (3, 3))
matrix_difference = matrix_c - matrix_d
print('Перша матриця:')
print(matrix_c)
print('Друга матриця:')
print(matrix_d)
print('Різниця матриць:')
print(matrix_difference)
```

```python
# 9. Створіть дві матриці розміром 3x3, заповніть їх випадковими цілими числами в діапазоні від -5 до 5 та знайдіть їх добуток.
matrix_e = np.random.randint(-5, 6, (3, 3))
matrix_f = np.random.randint(-5, 6, (3, 3))
matrix_product_ef = np.dot(matrix_e, matrix_f)
print('Перша матриця:')
print(matrix_e)
print('Друга матриця:')
print(matrix_f)
print('Добуток матриць:')
print(matrix_product_ef)
```

```python
# 10. Створіть матрицю розміром 3x3, заповніть її випадковими дійсними числами в діапазоні від -1 до 1 та знайдіть вектор-стовпчик, що містить суму елементів кожного рядка матриці.
matrix_sum_rows = np.random.uniform(-1, 1, (3, 3))
column_vector_sums = matrix_sum_rows.sum(axis=1).reshape(-1, 1)
print('Початкова матриця 3x3:')
print(matrix_sum_rows)
print('Вектор-стовпчик сум елементів кожного рядка:')
print(column_vector_sums)
```
