```python
df['month'] = pd.to_datetime(df['month'], format='%m/%d/%Y %I:%M:%S %p')

df_2018 = df[df['month'].dt.year == 2018]

min_visitors = df_2018.iloc[:, 1:].min()
max_visitors = df_2018.iloc[:, 1:].max()
result = pd.DataFrame({'min': min_visitors, 'max': max_visitors})

result
```

```python
# Фільтруємо лише за 2015 рік
df_2015 = df.loc[df['month'].dt.year == 2015].copy()

# Додаємо новий стовпець із загальною кількістю відвідувачів по всіх музеях
df_2015['total_visitors'] = df_2015.iloc[:, 1:].sum(axis=1)

# Знаходимо рядок з мінімальною та максимальною кількістю
min_row = df_2015.loc[df_2015['total_visitors'].idxmin()]
max_row = df_2015.loc[df_2015['total_visitors'].idxmax()]

# Виводимо назви місяців
min_month = min_row['month'].strftime('%B')
max_month = max_row['month'].strftime('%B')

print(f'Найнижча відвідуваність у {min_month}')
print(f'Найвища відвідуваність у {max_month}')
```

```python
df_2018 = df[df['month'].dt.year == 2018].copy()

# Витягаємо номер місяця
df_2018['month_num'] = df_2018['month'].dt.month

# Літні місяці: червень (6), липень (7), серпень (8)
summer = df_2018[df_2018['month_num'].isin([6, 7, 8])]['avila_adobe'].sum()

# Зимові місяці: грудень (12), січень (1), лютий (2)
# Потрібно об'єднати грудень 2017 та січень-лютий 2018 — але якщо лиш 2018, то:
winter = df_2018[df_2018['month_num'].isin([1, 2, 12])]['avila_adobe'].sum()

print(f" Літом відвідали: {summer} людей")
print(f" Зимою відвідали: {winter} людей")
```

```python
mtx = df.corr(numeric_only=True).abs()

# Побудова теплової карти кореляції
fig, ax = plt.subplots(figsize=(8, 8))

sns.heatmap(
    mtx,
    cmap='crest',
    annot=True,
    fmt=".2f",
    linewidths=0.5,
    mask=np.triu(np.ones_like(mtx, dtype=bool)),  # маскуємо верхній трикутник
    square=True,
    cbar=False,
    ax=ax
)

plt.title("Матриця попарної кореляції ознак", fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# 1. Лінійний графік (відвідуваність Avila Adobe з часом)
sns.set_theme(style="darkgrid")
sns.lineplot(data=df, x='month', y='avila_adobe')
plt.title('Avila Adobe: Відвідуваність з часом')
plt.show()
```

```python
# 3. Histogram (розподіл кількості відвідувачів Gateway to Nature Center)
sns.histplot(df['gateway_to_nature_center'], bins=30, kde=True)
plt.title('Розподіл відвідуваності Gateway to Nature Center')
plt.show()
```

```python
# 5. Barplot (середня кількість відвідувачів по музеях за всі роки)
avg_visitors = df.drop(columns=['month', 'year']).mean().sort_values(ascending=False)
sns.barplot(x=avg_visitors.values, y=avg_visitors.index)
plt.title('Середня кількість відвідувачів по музеях')
plt.xlabel('Середнє значення')
plt.ylabel('Музей')
plt.show()
```

```python
penguins_df = sns.load_dataset('penguins')
penguins_df
```

```python
penguins_df.info()
```

```python
penguins_df.isnull().sum()
```

```python
penguins_df = penguins_df.dropna().copy()

penguins_df['body_mass_kg'] = penguins_df['body_mass_g'] / 1000
penguins_df['bill_area'] = penguins_df['bill_length_mm'] * penguins_df['bill_depth_mm']
penguins_df['flipper_bill_ratio'] = penguins_df['flipper_length_mm'] / penguins_df['bill_length_mm']
penguins_df['mass_group'] = pd.qcut(
    penguins_df['body_mass_g'],
    q=3,
    labels=['light', 'medium', 'heavy']
)

penguins_df.head()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=penguins_df,
    x='bill_area',
    y='body_mass_kg',
    hue='species',
    style='sex',
    size='flipper_length_mm'
)

plt.title('Залежність маси тіла від площі дзьоба у пінгвінів')
plt.xlabel('Площа дзьоба')
plt.ylabel('Маса тіла, кг')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(
    data=penguins_df,
    x='species',
    y='flipper_bill_ratio',
    hue='sex',
    inner='quartile'
)

plt.title('Розподіл співвідношення довжини плавника до довжини дзьоба')
plt.xlabel('Вид пінгвіна')
plt.ylabel('Flipper / Bill ratio')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.swarmplot(
    data=penguins_df,
    x='island',
    y='bill_area',
    hue='species',
    size=5
)

plt.title('Розподіл площі дзьоба пінгвінів по островах')
plt.xlabel('Острів')
plt.ylabel('Площа дзьоба')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(
    data=penguins_df,
    x='bill_length_mm',
    y='bill_depth_mm',
    hue='species',
    fill=True
)

plt.title('Щільність розподілу довжини та глибини дзьоба')
plt.xlabel('Довжина дзьоба, мм')
plt.ylabel('Глибина дзьоба, мм')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.ecdfplot(
    data=penguins_df,
    x='body_mass_g',
    hue='species'
)

plt.title('Накопичувальний розподіл маси тіла пінгвінів')
plt.xlabel('Маса тіла, г')
plt.ylabel('Частка спостережень')
plt.show()
```
